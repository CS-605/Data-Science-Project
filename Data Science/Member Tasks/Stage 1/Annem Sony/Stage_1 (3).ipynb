```python
import pandas as pd
import matplotlib.pyplot as plt

october_employment_column = 'October Employment'
november_employment_column = 'November Employment'
december_employment_column = 'December Employment'

# Filter out rows with missing values in the selected columns
df = emp_merge.dropna(subset=[october_employment_column, november_employment_column, december_employment_column])

# Calculate the total count of employment and cases for each month
total_october_employment = df[october_employment_column].sum()
total_november_employment = df[november_employment_column].sum()
total_december_employment = df[december_employment_column].sum()

# Create a list of months and corresponding employment totals
months = ['October', 'November', 'December']
employment_totals = [total_october_employment, total_november_employment, total_december_employment]

# Define the cases columns for the corresponding months
cases_columns = ['2020-10-12_cases', '2020-11-12_cases', '2020-12-12_cases']

# Calculate the total count of cases for each month
total_cases = [df[cases_column].sum() for cases_column in cases_columns]

# Create a figure with one subplot
plt.figure(figsize=(10, 6))

# Plot total employment and cases for each month
plt.bar(months, employment_totals, color='lightblue', label='Employment')
plt.bar(months, total_cases, color='red', label='Cases', alpha=0.5)
plt.title('Total Employment and Cases for October, November, and December')
plt.xlabel('Month')
plt.ylabel('Count')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
