## Calculate COVID-19 data trends for last month of the data. Are the cases increasing, decreasing, or stable? Each student chooses a state to analyze

Description: In this task i will be computing the data trends for last month data. The data present in the covid cases dataset from usafacts.org do have 2023-07-23 as the last date, that is, the cases data after that specific date are not recorded. So, i am considering the last month date period as 2023-06-23 to 2023-07-23. And i have opted NC state for data analysis.

```python
#importing required packages
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading covid cases dataset into a dataframe
df_cases=pd.read_csv("../../../Consumed Datasets/Stage 1/covid_confirmed_usafacts.csv")
df_cases
```

```python
# Dropping unwanted columns from the cases dataset
df_cases=df_cases.drop(['countyFIPS','County Name','StateFIPS'],axis=1)
df_cases
```

```python
# Selecting a state to perform analysis
selected_state='NC'

#Filtering the dataset
state_cases= df_cases[df_cases['State']==selected_state]
state_cases
```

```python
# Grouping the NC state cases data.
state_cases = pd.DataFrame(state_cases.groupby(by='State').sum()).reset_index()
state_cases
```

```python
# Extracting the first column i.e., state column
first_column = state_cases.iloc[:, 0:1]

# Extracting the last 31 columns i.e., last month data
last_30_columns = state_cases.iloc[:, -31:]

# Combining them into a single DataFrame
last_month_nc_cases = pd.concat([first_column, last_30_columns], axis=1)
last_month_nc_cases
```

```python
# Dropping state column 
last_month_nc_cases=last_month_nc_cases.drop(['State'],axis=1)
last_month_nc_cases
```

```python
# Transorming data so that it is well suited for plotting graph.
last_month_nc_cases= pd.DataFrame(last_month_nc_cases.transpose()).reset_index()
last_month_nc_cases.columns=['Date','Cases']
last_month_nc_cases
```

```python
# Creating a plot for the last months data
plt.figure(figsize=(12, 6))
plt.plot(last_month_nc_cases.Date, last_month_nc_cases.Cases, marker='o', linestyle='-', color='r')
plt.title(f'Daily New COVID-19 Cases Trend in {selected_state} (June 23 2023 - July 23 2023)')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.xticks(rotation=45)
plt.show(True)
```

### SUMMARY
#### From the above graph we can conslude that the total covid cases for the NC state are neither increasing nor decreasing the count remianed constant  that is, "STABLE" through out the month.

#### If we need to calculate COVID-19 data trends for last month of the data which is the month of 'July' unlike one month peiod as above. We can do that by considering only those specific columns. Below are steps for doing so.

```python
# Extracting JULY's data
JULY_nc_cases= pd.DataFrame(last_month_nc_cases.tail(23)).reset_index()
JULY_nc_cases
```

```python
# Creating a plot for the last months data
plt.figure(figsize=(12, 6))
plt.plot(JULY_nc_cases.Date, JULY_nc_cases.Cases, marker='o', linestyle='-', color='r')
plt.title(f'Daily New COVID-19 Cases Trend in {selected_state} for the month of JULY')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.xticks(rotation=45)
plt.show(True)
```

As a part of this task i will be selecting the fields/ variables which are common in both the super covid dataset and enrichment dataset and perform the merge operation.

```python
# Merging the cases, deaths and population covid files to generate supercovid dataset
df_cases=pd.read_csv("../../../../Data Science/Consumed Datasets/Stage 1/covid_confirmed_usafacts.csv")
df_deaths=pd.read_csv("../../../../Data Science/Consumed Datasets/Stage 1/covid_deaths_usafacts.csv")
df_population=pd.read_csv("../../../../Data Science/Consumed Datasets/Stage 1/covid_county_population_usafacts.csv")
Merge_df = pd.merge(df_cases,df_deaths,on=('countyFIPS','County Name','State','StateFIPS'),how='outer')
df_population=df_population.drop(['County Name','State'],axis=1)
df_supercovid19=pd.merge(Merge_df,df_population, on=('countyFIPS'))
df_supercovid19
```
