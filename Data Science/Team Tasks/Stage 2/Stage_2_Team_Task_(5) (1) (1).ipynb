#### Task 3 Plot weekly trends (cases and deaths) of US and compare other countries. Utilize aggregrate, normalized by population, difference in cases, and log normalized values. (hint: do normalization: number of cases per 100,000, and then use np.log10())

### Convert the US dataset into suitable formate based on date for all states
Added state wise cases and deaths for each date, 184 days 27 weeks

```python
import pandas as pd

cases_deaths['date'] = pd.to_datetime(cases_deaths['date'])
start_date = pd.to_datetime('2020-07-01')
end_date = pd.to_datetime('2020-12-31')
date_list = pd.date_range(start=start_date, end=end_date, freq='D')
total_cases_list = []
total_deaths_list = []
total_population_list = []

for date in date_list:
    date_mask = (cases_deaths['date'] == date)
    total_cases = cases_deaths[date_mask]['cases'].sum()
    total_deaths = cases_deaths[date_mask]['deaths'].sum()
    total_population = cases_deaths[date_mask]['population'].sum()
    total_cases_list.append(total_cases)
    total_deaths_list.append(total_deaths)
    total_population_list.append(total_population)

US_cases_deaths_df = pd.DataFrame({
    'date': date_list,
    'total_cases': total_cases_list,
    'total_deaths': total_deaths_list,
    'total_population': total_population_list
})

print(US_cases_deaths_df)
```

### Method1: Using Aggregate operation in order to compare with other countries

```python
import pandas as pd
cases_deaths

# Assuming you have a DataFrame named 'cases_deaths' with 'date', 'cases', and 'deaths' columns

# Group by week and calculate the sum of cases and deaths
peak_weekly = cases_deaths.groupby([pd.Grouper(key='date', freq='W')]).agg({'cases': 'sum', 'deaths': 'sum'}).reset_index()
peak_weekly.columns = ['Week', 'Cases', 'Deaths']

# Find the row with the highest total cases
highest_cases_row = peak_weekly[peak_weekly['Cases'] == peak_weekly['Cases'].max()]
highest_cases_week = highest_cases_row.iloc[0]['Week']
highest_cases_count = highest_cases_row.iloc[0]['Cases']
highest_deaths_row =peak_weekly[peak_weekly['Deaths'] == peak_weekly['Deaths'].max()]
highest_deaths_week = highest_deaths_row.iloc[0]['Week']
highest_deaths_count = highest_deaths_row.iloc[0]['Deaths']

other_countries
india_data = other_countries[other_countries['location'] == 'India']
china_data = other_countries[other_countries['location'] == 'China']
indonesia_data = other_countries[other_countries['location'] == 'Indonesia']
pakistan_data = other_countries[other_countries['location'] == 'Pakistan']
nigeria_data = other_countries[other_countries['location'] == 'Nigeria']

india_weekly = india_data.groupby([pd.Grouper(key='date', freq='W')]).agg(
    india_cases=('total_cases', 'sum'),
    india_deaths=('total_deaths', 'sum')
).reset_index()
india_weekly.columns = [ 'Week','INDIA_Cases','INDIA_Deaths']

china_weekly = china_data.groupby([pd.Grouper(key='date', freq='W')]).agg(
    china_cases=('total_cases', 'sum'),
    china_deaths=('total_deaths', 'sum')
).reset_index()
china_weekly.columns = [ 'Week','China_Cases','China_Deaths']
indonesia_weekly = indonesia_data.groupby([pd.Grouper(key='date', freq='W')]).agg(
    indonesia_cases=('total_cases', 'sum'),
    indonesia_deaths=('total_deaths', 'sum')
).reset_index()
indonesia_weekly.columns = [ 'Week','indonesia_Cases','indonesia_Deaths']
pakistan_weekly = pakistan_data.groupby([pd.Grouper(key='date', freq='W')]).agg(
    pakistan_cases=('total_cases', 'sum'),
    pakistan_deaths=('total_deaths', 'sum')
).reset_index()
pakistan_weekly.columns = [ 'Week','pakistan_Cases','pakistan_Deaths']
nigeria_weekly = nigeria_data.groupby([pd.Grouper(key='date', freq='W')]).agg(
    nigeria_cases=('total_cases', 'sum'),
    nigeria_deaths=('total_deaths', 'sum')
).reset_index()
nigeria_weekly.columns = [ 'Week','nigeria_Cases','nigeria_Deaths']

highest_cases_row1 = india_weekly[india_weekly['INDIA_Cases'] == india_weekly['INDIA_Cases'].max()]
highest_cases_week1 = highest_cases_row1.iloc[0]['Week']
highest_cases_count1 = highest_cases_row1.iloc[0]['INDIA_Cases']
highest_deaths_row1 =india_weekly[india_weekly['INDIA_Deaths'] == india_weekly['INDIA_Deaths'].max()]
highest_deaths_week1 = highest_deaths_row1.iloc[0]['Week']
highest_deaths_count1 = highest_deaths_row1.iloc[0]['INDIA_Deaths']

highest_cases_row2 = china_weekly[china_weekly['China_Cases'] == china_weekly['China_Cases'].max()]
highest_cases_week2 = highest_cases_row2.iloc[0]['Week']
highest_cases_count2 = highest_cases_row2.iloc[0]['China_Cases']
highest_deaths_row2 =china_weekly[china_weekly['China_Deaths'] == china_weekly['China_Deaths'].max()]
highest_deaths_week2 = highest_deaths_row2.iloc[0]['Week']
highest_deaths_count2 = highest_deaths_row2.iloc[0]['China_Deaths']

highest_cases_row3 = indonesia_weekly[indonesia_weekly['indonesia_Cases'] == indonesia_weekly['indonesia_Cases'].max()]
highest_cases_week3 = highest_cases_row3.iloc[0]['Week']
highest_cases_count3 = highest_cases_row3.iloc[0]['indonesia_Cases']
highest_deaths_row3 =indonesia_weekly[indonesia_weekly['indonesia_Deaths'] == indonesia_weekly['indonesia_Deaths'].max()]
highest_deaths_week3 = highest_deaths_row3.iloc[0]['Week']
highest_deaths_count3 = highest_deaths_row3.iloc[0]['indonesia_Deaths']


highest_cases_row4 = pakistan_weekly[pakistan_weekly['pakistan_Cases'] == pakistan_weekly['pakistan_Cases'].max()]
highest_cases_week4 = highest_cases_row4.iloc[0]['Week']
highest_cases_count4 = highest_cases_row4.iloc[0]['pakistan_Cases']
highest_deaths_row4 =pakistan_weekly[pakistan_weekly['pakistan_Deaths'] == pakistan_weekly['pakistan_Deaths'].max()]
highest_deaths_week4 = highest_deaths_row4.iloc[0]['Week']
highest_deaths_count4 = highest_deaths_row4.iloc[0]['pakistan_Deaths']

highest_cases_row5 = nigeria_weekly[nigeria_weekly['nigeria_Cases'] == nigeria_weekly['nigeria_Cases'].max()]
highest_cases_week5 = highest_cases_row5.iloc[0]['Week']
highest_cases_count5 = highest_cases_row5.iloc[0]['nigeria_Cases']
highest_deaths_row5 =nigeria_weekly[nigeria_weekly['nigeria_Deaths'] == nigeria_weekly['nigeria_Deaths'].max()]
highest_deaths_week5 = highest_deaths_row5.iloc[0]['Week']
highest_deaths_count5 = highest_deaths_row5.iloc[0]['nigeria_Deaths']


# Create a dictionary to store the data
data = {
    'Country': ['US', 'India', 'China', 'Indonesia', 'Pakistan', 'Nigeria'],
    'Highest Cases Week': [highest_cases_week, highest_cases_week1, highest_cases_week2, highest_cases_week3, highest_cases_week4, highest_cases_week5],
    'Highest Cases Count': [highest_cases_count, highest_cases_count1, highest_cases_count2, highest_cases_count3, highest_cases_count4, highest_cases_count5],
    'Highest Deaths Week': [highest_deaths_week, highest_deaths_week1, highest_deaths_week2, highest_deaths_week3, highest_deaths_week4, highest_deaths_week5],
    'Highest Deaths Count': [highest_deaths_count, highest_deaths_count1, highest_deaths_count2, highest_deaths_count3, highest_deaths_count4, highest_deaths_count5]
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Print the DataFrame
df
```
